```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8')

print("✅ All libraries imported successfully!")
```

```python
# Load the light version of the dataset
df = pd.read_json('../data/houston_housing market 2024_light.json')

print("✅ Data loaded!")
print(f"Shape: {df.shape}")
print(f"\nColumns:\n{df.columns.tolist()}")
```

```python
df.head()
```

```python
# Keep only the columns we need
cols = ['unformattedPrice', 'beds', 'baths', 'area', 
        'addressZipcode', 'zestimate', 'statusType', 'brokerName']
df = df[cols]

# Rename for clarity
df.columns = ['price', 'beds', 'baths', 'area_sqft', 
              'zipcode', 'zestimate', 'status', 'broker']

# Drop rows with missing price, beds, baths, area
df = df.dropna(subset=['price', 'beds', 'baths', 'area_sqft'])

# Remove unrealistic values
df = df[df['price'] > 10000]
df = df[df['area_sqft'] > 100]
df = df[df['beds'] > 0]

# Reset index
df = df.reset_index(drop=True)

print(f"✅ Clean dataset shape: {df.shape}")
print(f"\nMissing values:\n{df.isnull().sum()}")
print(f"\nBasic Stats:")
print(df[['price', 'beds', 'baths', 'area_sqft']].describe())
```

```python
# Add a new calculated column
df['price_per_sqft'] = round(df['price'] / df['area_sqft'], 2)

print("✅ price_per_sqft column added!")
df.head()
```

```python
# Check missing values
print("Missing values BEFORE:\n", df.isnull().sum())

# 1. Zestimate — fill with median (it's a numeric column)
df['zestimate'] = df['zestimate'].fillna(df['zestimate'].median())

# 2. Broker — fill with 'Unknown' (it's a text column)
df['broker'] = df['broker'].fillna('Unknown')

# Verify
print("\nMissing values AFTER:\n", df.isnull().sum())
print("\n✅ All missing values handled!")
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return lower, upper

# Calculate IQR upper bounds
_, price_upper = remove_outliers_iqr(df, 'price')
_, area_upper = remove_outliers_iqr(df, 'area_sqft')
_, baths_upper = remove_outliers_iqr(df, 'baths')
_, beds_upper = remove_outliers_iqr(df, 'beds')

# Apply IQR upper bounds + logical lower bounds
df_clean = df[
    (df['price'] >= 50_000) & (df['price'] <= price_upper) &
    (df['area_sqft'] >= 300) & (df['area_sqft'] <= area_upper) &
    (df['baths'] >= 1) & (df['baths'] <= baths_upper) &
    (df['beds'] >= 1) & (df['beds'] <= beds_upper)
].copy().reset_index(drop=True)

print("Final Bounds Applied:")
print(f"Price:    $50,000 — ${price_upper:,.0f}")
print(f"Area:     300 — {area_upper:,.0f} sqft")
print(f"Baths:    1 — {baths_upper:.0f}")
print(f"Beds:     1 — {beds_upper:.0f}")
print(f"\nBEFORE: {df.shape[0]:,} rows")
print(f"AFTER:  {df_clean.shape[0]:,} rows")
print(f"Removed: {df.shape[0] - df_clean.shape[0]:,} rows")
print("\n✅ Outliers removed using IQR + domain knowledge!")
```

```python
print("=" * 50)
print("FINAL CLEAN DATASET SUMMARY")
print("=" * 50)
print(f"\nShape: {df_clean.shape}")
print(f"\nData Types:\n{df_clean.dtypes}")
print(f"\nMissing Values:\n{df_clean.isnull().sum()}")
print(f"\nStatistical Summary:")
print(df_clean[['price', 'beds', 'baths', 'area_sqft', 'price_per_sqft']].describe().round(2))
print("\n✅ Dataset is clean and ready for analysis!")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Left plot — Price Distribution
axes[0].hist(df_clean['price'], bins=50, 
             color='steelblue', edgecolor='white', linewidth=0.5)
axes[0].set_title('Houston Home Price Distribution (2024)', 
                  fontsize=14, fontweight='bold', pad=15)
axes[0].set_xlabel('Price ($)', fontsize=12)
axes[0].set_ylabel('Number of Homes', fontsize=12)
axes[0].xaxis.set_major_formatter(
    plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))
axes[0].axvline(df_clean['price'].mean(), color='red', 
                linestyle='--', linewidth=1.5, label=f"Mean: ${df_clean['price'].mean():,.0f}")
axes[0].axvline(df_clean['price'].median(), color='green', 
                linestyle='--', linewidth=1.5, label=f"Median: ${df_clean['price'].median():,.0f}")
axes[0].legend(fontsize=10)

# Right plot — Price per Sqft Distribution
axes[1].hist(df_clean['price_per_sqft'], bins=50, 
             color='darkorange', edgecolor='white', linewidth=0.5)
axes[1].set_title('Price per Sqft Distribution (2024)', 
                  fontsize=14, fontweight='bold', pad=15)
axes[1].set_xlabel('Price per Sqft ($)', fontsize=12)
axes[1].set_ylabel('Number of Homes', fontsize=12)
axes[1].axvline(df_clean['price_per_sqft'].mean(), color='red', 
                linestyle='--', linewidth=1.5, label=f"Mean: ${df_clean['price_per_sqft'].mean():.2f}")
axes[1].axvline(df_clean['price_per_sqft'].median(), color='green', 
                linestyle='--', linewidth=1.5, label=f"Median: ${df_clean['price_per_sqft'].median():.2f}")
axes[1].legend(fontsize=10)

plt.suptitle('Houston Housing Market — Price Analysis', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../visuals/01_price_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Visualization 1 saved!")
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

scatter = ax.scatter(
    df_clean['area_sqft'],
    df_clean['price'],
    c=df_clean['beds'],
    cmap='viridis',
    alpha=0.4,
    s=15
)

cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Number of Bedrooms', fontsize=11)
cbar.set_ticks([1, 2, 3, 4, 5])

# Add trend line
z = np.polyfit(df_clean['area_sqft'], df_clean['price'], 1)
p = np.poly1d(z)
x_line = np.linspace(df_clean['area_sqft'].min(), df_clean['area_sqft'].max(), 100)
ax.plot(x_line, p(x_line), color='red', linewidth=2, linestyle='--', label='Trend Line')

ax.set_title('Home Price vs Area — Houston 2024', 
             fontsize=15, fontweight='bold', pad=15)
ax.set_xlabel('Area (sqft)', fontsize=12)
ax.set_ylabel('Price ($)', fontsize=12)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))
ax.legend(fontsize=11)

plt.tight_layout()
plt.savefig('../visuals/02_price_vs_area.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Visualization 2 saved!")
```

```python
# Calculate median price and count per bedroom
bed_stats = df_clean.groupby('beds').agg(
    median_price=('price', 'median'),
    count=('price', 'count')
).reset_index()

fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart — median price
bars = ax1.bar(bed_stats['beds'].astype(int),
               bed_stats['median_price'],
               color='steelblue', edgecolor='white',
               linewidth=0.5, width=0.5, label='Median Price')

# Add price labels on top of each bar
for bar, val in zip(bars, bed_stats['median_price']):
    ax1.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 3000,
             f'${val:,.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Second y-axis — listing count
ax2 = ax1.twinx()
ax2.plot(bed_stats['beds'].astype(int),
         bed_stats['count'],
         color='darkorange', marker='o',
         linewidth=2, markersize=7, label='Number of Listings')
ax2.set_ylabel('Number of Listings', fontsize=12, color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

ax1.set_title('Median Home Price & Listing Count by Bedrooms — Houston 2024',
              fontsize=14, fontweight='bold', pad=15)
ax1.set_xlabel('Number of Bedrooms', fontsize=12)
ax1.set_ylabel('Median Price ($)', fontsize=12)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Combined legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, fontsize=10, loc='upper left')

plt.tight_layout()
plt.savefig('../visuals/03_price_by_bedrooms.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Visualization 3 saved!")
```

```python
# Get top 15 zipcodes by listing count
top_zips = df_clean['zipcode'].value_counts().head(15).index
df_zip = df_clean[df_clean['zipcode'].isin(top_zips)]

# Calculate median price per sqft per zipcode
zip_stats = df_zip.groupby('zipcode').agg(
    median_ppsf=('price_per_sqft', 'median'),
    count=('price_per_sqft', 'count')
).reset_index().sort_values('median_ppsf', ascending=True)

fig, ax = plt.subplots(figsize=(12, 8))

bars = ax.barh(zip_stats['zipcode'].astype(str),
               zip_stats['median_ppsf'],
               color='steelblue', edgecolor='white', linewidth=0.5)

# Add value labels
for bar, val in zip(bars, zip_stats['median_ppsf']):
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2,
            f'${val:.0f}',
            va='center', fontsize=10, fontweight='bold')

ax.set_title('Median Price per Sqft by Zipcode — Top 15 Houston Areas (2024)',
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Median Price per Sqft ($)', fontsize=12)
ax.set_ylabel('Zipcode', fontsize=12)
ax.set_xlim(0, zip_stats['median_ppsf'].max() + 30)

plt.tight_layout()
plt.savefig('../visuals/04_price_per_sqft_by_zipcode.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Visualization 4 saved!")
```

```python
# Select only numeric columns relevant to analysis
corr_cols = ['price', 'beds', 'baths', 'area_sqft', 
             'zestimate', 'price_per_sqft']

# Calculate correlation matrix
corr_matrix = df_clean[corr_cols].corr().round(2)

fig, ax = plt.subplots(figsize=(10, 8))

# Plot heatmap
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    vmin=-1, vmax=1,
    linewidths=0.5,
    linecolor='white',
    annot_kws={'size': 12, 'weight': 'bold'},
    ax=ax
)

ax.set_title('Feature Correlation Heatmap — Houston Housing 2024',
             fontsize=14, fontweight='bold', pad=15)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=11)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=11)

plt.tight_layout()
plt.savefig('../visuals/05_correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Visualization 5 saved!")
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

# Get data per bedroom group
bed_groups = [df_clean[df_clean['beds'] == b]['price'].values 
              for b in sorted(df_clean['beds'].unique())]
bed_labels = [f'{int(b)} Bed' for b in sorted(df_clean['beds'].unique())]

bp = ax.boxplot(
    bed_groups,
    labels=bed_labels,
    patch_artist=True,
    medianprops=dict(color='white', linewidth=2),
    flierprops=dict(marker='o', markerfacecolor='gray', 
                    markersize=3, alpha=0.3),
    whiskerprops=dict(linewidth=1.5),
    capprops=dict(linewidth=1.5)
)

# Color each box
colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#CCB974']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.8)

ax.set_title('Price Distribution by Number of Bedrooms — Houston 2024',
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Bedroom Category', fontsize=12)
ax.set_ylabel('Price ($)', fontsize=12)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Add median labels
for i, group in enumerate(bed_groups):
    median = np.median(group)
    ax.text(i + 1, median + 8000, f'${median/1000:.0f}K',
            ha='center', fontsize=9, fontweight='bold', color='black')

plt.tight_layout()
plt.savefig('../visuals/06_price_boxplot_by_bedrooms.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Visualization 6 saved!")
```

```python
# Calculate listing status counts and percentages
status_counts = df_clean['status'].value_counts()
status_pct = (status_counts / status_counts.sum() * 100).round(1)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Left — Bar chart
colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#CCB974'][:len(status_counts)]
bars = axes[0].bar(status_counts.index, status_counts.values,
                   color=colors, edgecolor='white', linewidth=0.5)

for bar, val, pct in zip(bars, status_counts.values, status_pct.values):
    axes[0].text(bar.get_x() + bar.get_width() / 2,
                 bar.get_height() + 100,
                 f'{val:,}\n({pct}%)',
                 ha='center', va='bottom', fontsize=10, fontweight='bold')

axes[0].set_title('Listing Count by Status Type', fontsize=14, fontweight='bold', pad=15)
axes[0].set_xlabel('Status Type', fontsize=12)
axes[0].set_ylabel('Number of Listings', fontsize=12)
axes[0].set_xticklabels(status_counts.index, rotation=15, ha='right')

# Right — Pie chart
wedges, texts, autotexts = axes[1].pie(
    status_counts.values,
    labels=status_counts.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    wedgeprops=dict(edgecolor='white', linewidth=1.5)
)
for text in autotexts:
    text.set_fontsize(11)
    text.set_fontweight('bold')

axes[1].set_title('Market Share by Status Type', fontsize=14, fontweight='bold', pad=15)

plt.suptitle('Houston Housing Market — Listing Status Analysis 2024',
             fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../visuals/07_listing_status.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Visualization 7 saved!")
```
